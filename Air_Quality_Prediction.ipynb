```python
air_quality_data_part1 = air_quality_data.iloc[:,:2]  # First 2 rows
air_quality_data_part2 = air_quality_data.iloc[:,2:]
```

```python
air_quality_data_part2 = air_quality_data_part2.fillna(air_quality_data_part2.mean())
```

```python
air_quality_data_part2.tail()
```

```python
air_quality_data = pd.concat([air_quality_data_part1, air_quality_data_part2], axis=1)
```

```python
air_quality_data.tail()
```

```python
#Forecasting with FB Prophet Algorithm
#before forecating we need to convert date from DD/MM/YYYY format to YYYY-MM-DD

date_info = pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y')
```

```python
print(date_info)
```

```python
time_info = air_quality_data['Time']
time_info = time_info.apply(lambda x:x.replace('.',':'))
```

```python
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
date_time = pd.concat([date_info, time_info], axis = 1)
```

```python
date_time.head()
```

```python
date_time.shape
```

```python
#combining date and time
date_time['ds'] = date_time['Date'].astype(str) +' ' + date_time['Time'].astype(str)
```

```python
date_time.head()
```

```python
date_time.info()
```

```python
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y'] = air_quality_data['RH']
```

```python
data.head()
```

```python
from prophet import Prophet
```

```python
model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
